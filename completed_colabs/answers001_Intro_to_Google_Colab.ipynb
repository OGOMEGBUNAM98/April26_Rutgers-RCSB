## Section 5 — Runtime Type

A **runtime** is the computing environment that actually runs your code. Colab gives you a choice of hardware and coding language (Python, R, or Julia).

The type of hardware you need will vary. For small datasets (less than 100,000 rows of data), fitting enzyme kinetics, visualizing most BMB-related plots, a CPU (Central Processing Unit) runtime is fine. For molecular dynamics or running protein structure modeling, a GPU (Graphics Processing Unit) runtime might be needed. Also, not all Python packages can run on a GPU, so no need to enable what you don't need.

The Free Colab plan limits the connection time and types of hardware available, paid plans increase your options. Using a higher end GPU reduces your connection time.

> For most activities in BMB, the free Colab plan is sufficient.

**How to set your runtime type:**
1. In the top menu, click **Runtime → Change runtime type**.
2. A dialog box will appear with a **Hardware accelerator** dropdown.
3. Choose one of:
   - **CPU** — Standard option for most tasks (default).
   - **T4 GPU** — For molecular dynamics or structure prediction with limits (free tier available).
   - **TPU** — Tensor Processing Unit, for advanced deep learning.
4. Click **Save**.

> **Note:** Changing the runtime type will disconnect and restart your session, clearing any data.

## Section 11 — Filling Text Boxes (Forms)

A unique feature of Google Colab is the ability to make the Colab formatted like a form that runs code. This can be helpful for complex processes or for teaching new coders to develop comfort. In these forms, instead of editing code directly, users can fill in text boxes, dropdowns, and sliders.

Form fields are created with special comments in code cells using `#@param`.

**Common form field types:**
- `#@param {type:"string"}` — Text input box
- `#@param ["option1", "option2"]` — Dropdown menu
- `#@param {type:"integer"}` — Number input
- `#@param {type:"boolean"}` — Checkbox

Try filling in the form fields in the cell below and then running it! Then click the `Show code` link to see the code running behind the form.

---

````python
#@title Tell Us About Yourself!

your_name = "Chris Berndsen" #@param {type:"string"}
your_experience = "Intermediate" #@param ["Beginner", "Intermediate", "Advanced"]
your_age = 40 #@param {type:"integer"}
enjoys_coding = True #@param {type:"boolean"}

# Now use the filled-in values
print(f"👋 Hello, {your_name}!")
print(f"📊 Experience level: {your_experience}")
print(f"🎂 Age: {your_age}")
print(f"💻 Enjoys coding: {enjoys_coding}")
print()
print(f"Welcome to Colab, {your_name}! You're going to do great things.")
````

---
#### Thoughts to consider when using AI to help code

Gemini (and other similar LLMs) do not fully know what you want the code to do or who is using the code. *Even for the simplest of inquiries, the AI may not provide a clear or efficient answer.* It is important to understand working code and what is happening.

**Some tips for using AI for coding:**

- Pre-outline your code steps or what you want the students to do.  This will help you write specific prompts.
- Build one step at a time. Asking for a whole series steps may result in code that can achieve the goal but is challenging to explain.
- For common packages with good documentation (ex. Pandas, Numpy, Scipy), Gemini works well. For discipline specific packages, you will not get great answers as the LLM needs a lot of data/text to scrape when developing answers.
- Test the AI generated code many times and with different inputs (or have your students test the code.) Frequently, AI generated code designs it around an existing data file in the Colab and does not generalize it to all data files you might want to use. Common issues include using specific column names or row numbers.
